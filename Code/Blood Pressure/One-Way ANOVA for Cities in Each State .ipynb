```python
df1 = pd.read_csv('SuperFilteredData.csv') # Data File imported from excel with all filtration done beforehand
cali_city = df1.iloc[1:5482]
```

```python
california = ols('Data_Value ~ CityName', data=cali_city).fit()  # Simple anova table based on our data, showing only the the F test statistic, and Probability value for States
aov_table= sm.stats.anova(california, typ=2) # Python's Anova version
print (aov_table) # Simplified Anova table Results
california.summary()
```

```python
floridafile = pd.read_csv('florida.csv')
```

```python
florida = ols('Data_Value ~ CityName', data=floridafile).fit()  # Simple anova table based on our data, showing only the the F test statistic, and Probability value for States
aov_table= sm.stats.anova_lm(florida, typ=2) # Python's Anova version
print (aov_table) # Simplified Anova table Results
florida.summary()
```

```python
massachusettsfile = pd.read_csv('massachusetts.csv')
```

```python
massachusetts = ols('Data_Value ~ CityName', data=massachusettsfile).fit()  # Simple anova table based on our data, showing only the the F test statistic, and Probability value for States
aov_table= sm.stats.anova_lm(massachusetts, typ=2) # Python's Anova version
print (aov_table) # Simplified Anova table Results
massachusetts.summary()
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd, MultiComparison #DifFerences between groups on one way ANOVA
mc = MultiComparison(cali_city['Data_Value'],cali_city['CityName'])
result=mc.tukeyhsd()
print(result)
```
